# Kicking horses

In 1898, the russian statistician Ladislaus Bortkiewicz studied the distribution of 122 men kicked to death by horses among 10 Prussian army corps within 20 years (so a total number of corpes of 200). He recorded the number of corps with a given number of deaths:

| Number of deaths | Number of groups |
| --- | --- |
| 0 | 109 |
| 1 | 65 |
| 2 | 22 |
| 3 | 3 |
| 4 | 1 |

### Let's plot the resulting probability distribution

```python
import numpy as np
import scipy.stats
import random
from scipy.stats import poisson
from matplotlib import pyplot as plt
```

```python
deaths = np.array([0,1,2,3,4])
groups = np.array([109,65,22,3,1])

tot = np.sum(groups)
freq = groups/tot


plt.scatter(deaths, freq, color='turquoise')
plt.title('Frequency of death among an army corp')
plt.ylabel('Frequency')
plt.xlabel('Number of deaths')
plt.show()
```

#### We expect a Poisson distribution since the death by an horse kick can be classified as "rare event"
Let's plot several poissonian with different $\mu$.

```python
plt.scatter(deaths, freq, color='turquoise')

for mu in [0.1,0.5,1,2]:
    p = scipy.stats.poisson(mu).pmf(deaths)
    plt.plot(deaths, p ,label='$\mu='+str(mu)+"$")
    
plt.title('Frequency of death among an army corp')
plt.ylabel('Frequency')
plt.xlabel('Number of deaths')
plt.legend()
```

#### Now we try to compute a fit by hand

```python
goodmu = np.average(deaths,weights=freq)
print("Estimate of Î¼: %.2f" %goodmu)
plt.scatter(deaths, freq, color='turquoise')

p = scipy.stats.poisson(goodmu).pmf(deaths)
plt.plot(deaths, p,linewidth=2, label='Fit', color='orange') 
plt.title('Fit of the data')
plt.ylabel('Frequency')
plt.xlabel('Number of deaths')
plt.legend()
```
