```python
import numpy as np
import matplotlib.pyplot as plt

N = 10000
x = np.random.uniform(0.1,10, N)

plt.hist(x, bins=20)
```

```python
y = np.log10(x)

def p_y(y):
    return (1/9.9)*np.log(10)*pow(10,y)

Y = np.sort(y)

plt.hist(y, density=True, bins=20)
plt.plot(Y, p_y(Y))
plt.show()
```

```python
#Computing

mean_x = np.mean(x)
mean_y = np.mean(y)
log_mean_x = np.log10 (mean_x)

print(mean_x)
print(log_mean_x)
print(mean_y)
```

```python
import statistics


median_x = statistics.median(x)
median_y = statistics.median(y)
log_median_x = np.log10 (median_x)

print(median_x)
print(log_median_x)
print(median_y)
```
